```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Customer Preference Survey (Responses) - Form Responses 1.csv')
print(df.shape)
df.head(5)
```

```python
df.isnull().sum()
```

```python
for i in df.columns[1:]:
    print(i, df[i].unique())
```

```python
df_transform = df.copy()
# df_transform.dropna(inplace=True)
df_transform.fillna(0, inplace=True)
# dataset น้อย ขอถือว่า null เป็นไม่เคยซื้อ
df_transform.replace('ไม่เคยซื้อ', 0, inplace=True)
df_transform.replace('ไม่เคยซือ', 0, inplace=True)
df_transform.replace('ไม่เคย', 0, inplace=True)
df_transform.replace('ไม่', 0, inplace=True)
df_transform.replace('เคย', 1, inplace=True)
df_transform.replace('เคยซื้อ', 1, inplace=True)
print(df_transform.shape)
df_transform.head(5)
```

```python
for i in df_transform.columns[1:]:
    print(i, df_transform[i].unique())
```

```python
dict_count = df_transform.iloc[:,1:].sum().to_dict()
dict_count
```

```python
sort_orders = sorted(dict_count.items(), key=lambda x: x[1], reverse=True)

df_top = pd.DataFrame(sort_orders, columns=['item','count'])
df_top.head(10)
```

```python
ax = df_top[:10].plot.bar(x='item', y='count')
# ax = sns.barplot(x='item', y='count', data=df_top[:10])
```

```python
df_transform = df_transform.drop(['Timestamp'], axis=1)
```

```python
from mlxtend.frequent_patterns import association_rules, apriori

frequent_itemsets = apriori(df_transform, min_support=0.2, max_len = 4, use_colnames=True)
# limit item length to be 4
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
print(rules.shape)
rules.head(10)
```

```python
# find threshold
perc =[.20, .40, .50, .60, .65, .7, .75, .80, .90]
rules[['support']].describe(percentiles=perc).T
```

```python
selected_rule = rules[rules['support'] >= 0.26]
print(selected_rule.shape)
selected_rule.head(10)
```

```python
# find threshold
perc =[.20, .40, .50, .60, .65, .7, .75, .80, .90]
df_confidence = selected_rule[['confidence']].describe(percentiles=perc).T
df_confidence
```

```python
# find threshold
perc =[.20, .40, .50, .60, .65, .7, .75, .80, .90]
df_lift = selected_rule[['lift']].describe(percentiles=perc).T
df_lift
```

```python
confidence_threshold = df_confidence['80%'].tolist()[0]
lift_threshold = df_lift['80%'].tolist()[0]
print(confidence_threshold)
print(lift_threshold)
```

```python
summarize = selected_rule[ (selected_rule['confidence'] >= confidence_threshold) & (selected_rule['lift'] >= lift_threshold)].reset_index(drop=True)
summarize[:10]
```

```python
summarize.sort_values(by=['support'], ascending=False)[:10]
```

```python
summarize.sort_values(by=['lift'], ascending=False)[:10]
```

```python
summarize.sort_values(by=['confidence'], ascending=False)[:10]
```

## Summary
- มีการ fix ว่า item ที่จะนำมาใช้ max length จะไม่เกิน 4 item
- ถ้าดูจาก lift score จะได้ว่า item ที่มันซื้อมักจะเป็นของที่ใช้ภายในบ้าน เช่น หม้อทอด หรือ external harddisk แล้วจะมีการซื้อเครื่องซักผ้า
- ถ้าดูจาก confidence score จะดูมีความ common sense มากกว่า อย่างเช่น (ขนมจีนน้ำยาปู, พลาสเตอร์บรรเทาปวด ตราเสือ) คู่กับ (ยาดม, Salmon Sashimi) จะเห็นว่าพลาสเตอร์เป็นยา มีการซื้อ ยาดมด้วย ส่วน ขนมจีนน้ำยาปูก้จะมีการซื้อ sashimi ที่เป็นอาหาร
